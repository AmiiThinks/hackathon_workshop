## Logistic Regression

- Models the probability an object belongs to a class
- Values ranges from 0 to 1
- Can use threshold to classify into which classes a class belongs
- An S-shaped curve

$
\begin{align}
\sigma(t) = \frac{1}{1 + e^{-t}}
\end{align}
$


```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import math
```

### How does the Logistic Regression function looks like

```python
x = [i for i in range(-10, 10)]
y = [1.0/(1+math.exp(-1*i)) for i in x]

ax = plt.figure()
plt.plot(x, y,  color='black')
plt.axhline(0.5)
plt.show()
```

## Problem with the previous approach

- Using the complete dataset for training
- A small set of data should be a holdout dataset on which a model can be tested
- The __holdout set__ will say how good your model is w.r.t. unknown data - __Test Data__
